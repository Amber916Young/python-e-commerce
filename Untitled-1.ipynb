```python
import pandas as pd
```

```python
#读表，根据sheet名
df1 = pd.read_excel('AAAA.xlsx', sheet_name='table1')  
df2 = pd.read_excel('AAAA.xlsx', sheet_name='table2')  
df3 = pd.read_excel('AAAA.xlsx', sheet_name='table3')  
```

```python
# 只是检查有没有重复数据，不影响结果
duplicates = df3['BARCODE_CN'].duplicated()
duplicate_rows = df3[duplicates]
print("BARCODE_CN重复的数据 ")
print(len(duplicate_rows))
```

```python
#检查一下，默认打印五条数据
df1.head()
```

```python
df3.head(10) #打印十条数据
```

```python

merged_df = pd.merge(df3, df1[['barcode', 'inv']], left_on='BARCODE_CN', right_on='barcode', how='left')
merged_df.head()
# 删除多余的 "barcode "列
merged_df = merged_df.drop(columns=['barcode'])
# 清理数据，没有值的设置为0
# excel里面的命令-->   =IFERROR(VLOOKUP([@[BARCODE_CN]], $K:$L, 12, FALSE), "0")
merged_df['inv'] = merged_df['inv'].fillna(0)
# 导出 做 privot分析
merged_df.to_excel('merged_result.xlsx', index=False)
```

```python
# 检查 只查看库存不为0的行，不影响结果
filtered_df = merged_df[merged_df['inv'] != 0]
filtered_df.head(10)
```

```python

# 按品牌和季节分组，计算库存总和
summary_df = merged_df.groupby(['BRANDS', 'SEASON'])['inv'].sum().reset_index()
print("每个品牌、每个季节的库存汇总:")
summary_df.head(10)
```

```python
summary_df2 = merged_df.groupby(['BRANDS', 'FAMILY_EN'])['inv'].sum().reset_index()
print("每个品牌和 FAMILY_EN 的库存汇总：")
print(summary_df2)
```

```python
#定位到每款REF&COLOR,库存大于200件的有多少个款
grouped_df = merged_df.groupby(['REF', 'COLOR_CN'])['inv'].sum().reset_index()
filtered_df3 = grouped_df[grouped_df['inv'] > 200]
models_count_per_group = filtered_df3.groupby(['REF', 'COLOR_CN']).size().reset_index(name='model_count')
print(" > 200:")
models_count_per_group.head(469)
```
